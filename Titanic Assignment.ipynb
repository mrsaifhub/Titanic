```python
# Importing all libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler as mms
from sklearn.linear_model import LogisticRegression as lr

e = 2.718
```

```python
# Loading the data from the local file titanic_train

train = pd.read_csv('titanic_train.csv')
```

```python
train
```

```python
# cleaning the data and removing unnecessary features

train.drop(['Name','Age','Ticket','Cabin'],axis = 1,inplace = True)
```

```python
train
```

```python
# Converting string data to numeric

train.Sex = train.Sex.replace('female',0)
train.Sex = train.Sex.replace('male',1)
train.Embarked = train.Embarked.replace(np.nan,'S')
train.Embarked = train.Embarked.replace('S',0)
train.Embarked = train.Embarked.replace('C',1)
train.Embarked = train.Embarked.replace('Q',2)
```

```python
train
```

```python
# Converting data frame to numpy for clasfication

data = train.to_numpy()
x = data[:,:-1]
y = data[:,-1]
```

```python
# Scaling the data for better results
scaler = mms()
scaler.fit(x)
x = scaler.transform(x)
```

```python
# training the model 

alg = lr(C=2,solver ='sag')
alg.fit(x,y)
```

```python
# Loading the testing data cleaning and changing the data

test_data = pd.read_csv('titanic_test.csv')
test_data.drop(['Name','Age','Ticket','Cabin'],axis = 1,inplace = True)
test_data.Sex = test_data.Sex.replace('female',0)
test_data.Sex = test_data.Sex.replace('male',1)
test_data.Embarked = test_data.Embarked.replace(np.nan,'S')
test_data.Embarked = test_data.Embarked.replace('S',0)
test_data.Embarked = test_data.Embarked.replace('C',1)
test_data.Embarked = test_data.Embarked.replace('Q',2)
```

```python
# Converting to numpy to predict

test = test_data.to_numpy()
```

```python
# Getting test results and saving

test_results = alg.predict(test)

np.savetxt("titanic_results.csv",test_results,delimiter=",",fmt='%1.5f')
```
